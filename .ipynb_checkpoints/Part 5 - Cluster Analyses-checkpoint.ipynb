## Initial Set-Up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('working_data.csv', parse_dates=[0], infer_datetime_format= True)
```

```python
len(data)
```

```python
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

data['day_name'] = pd.Categorical(data['day_name'], categories=cats, ordered=True)
```

```python
cats = ['January', 'February', 'March', 'April', 'May', 'June', 
       'July', 'August', 'September', 'October', 'November', 'December']

data['month_name'] = pd.Categorical(data['month_name'], categories=cats, ordered=True)
```

## Clustering In Two Dimensions

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x = data[['sessions_stand', 'tts_stand']].values
```

```python
WCSS = []
for i in range(1,11):
    model = KMeans(n_clusters = i, init = 'k-means++', n_init = 10, random_state = 42)
    model.fit(x)
    WCSS.append(model.inertia_)

fig = plt.figure(figsize = (10, 10))
plt.plot(range(1,11), WCSS, linewidth=4, markersize=8, marker='o', color = '#377FA6')
plt.xticks(np.arange(11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 4, init = "k-means++", max_iter = 300, n_init = 10, random_state = 42)
y_clusters = model.fit_predict(x)
```

```python
sns.countplot(x = y_clusters, palette = 'YlGnBu')
```

```python
plt.figure(figsize = (20,10))
plt.scatter(x[y_clusters == 0,0],x[y_clusters == 0,1],s = 50, c = '#C9EEAF', label = "Cluster 1")
plt.scatter(x[y_clusters == 1,0],x[y_clusters == 1,1],s = 50, c = '#377FA6', label = "Cluster 2")
plt.scatter(x[y_clusters == 2,0],x[y_clusters == 2,1],s = 50, c = '#6EBBAD', label = "Cluster 3")
plt.scatter(x[y_clusters == 3,0],x[y_clusters == 3,1],s = 50, c = '#393487', label = "Cluster 4")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], s = 100, c = "black", label = "centroids")
plt.xlabel("Sessions -- >")
plt.ylabel("TTS -- >")
plt.legend()
plt.show()
```

#### Freq of Devices In Each Cluster

```python
data.groupby(['device_id', 'model_predict', 'model_descrip'])['sessions'].count().reset_index().sort_values(by=['sessions'], ascending=[False]).head()
```

#### Freq of Apps in Each Cluster

```python
app_freq = data.groupby(['app_name', 'model_predict', 'model_descrip']
                       )['device_id'].count().reset_index().sort_values(by=['model_descrip', 'device_id'],
                                                                        ascending=[True, False]) 

temp = app_freq.groupby('model_predict')['device_id'].sum().reset_index()

app_freq = pd.merge(app_freq, temp, how='left', on='model_predict', copy=False)

app_freq['percent'] = (app_freq['device_id_x']/app_freq['device_id_y']) * 100

app_freq = app_freq.rename(columns={'device_id_x': 'unique devices', 'device_id_y': 'total devices'})

app_freq.head()
```

```python
app_freq.to_html('FreqAppsPerCluster.html')
```

#### Sum Sessions For Each App In Each Cluster

```python
app_freq = data.groupby(['app_name', 'model_predict', 'model_descrip']
                       )['sessions'].sum().reset_index().sort_values(by=['model_descrip', 'sessions'], 
                                                                      ascending=False) 

temp = app_freq.groupby('model_predict')['sessions'].sum().reset_index()

app_freq = pd.merge(app_freq, temp, how='left', on='model_predict', copy=False)

app_freq['percent'] = (app_freq['sessions_x']/app_freq['sessions_y']) * 100

app_freq = app_freq.rename(columns={'sessions_x': 'Sum App Sessions', 'sessions_y': 'Total Cat Sessions'})

app_freq.head()
```

```python
app_freq.to_html('SumSessionsAppsPerCluster.html')
```

#### Average TTS For Each App in Each Cluster

```python
app_freq = data.groupby(['app_name', 'model_predict', 'model_descrip']
                       )['tts'].sum().reset_index().sort_values(by=['model_descrip', 'tts'], 
                                                                      ascending=False) 

temp = app_freq.groupby('model_predict')['tts'].sum().reset_index()

app_freq = pd.merge(app_freq, temp, how='left', on='model_predict', copy=False)

app_freq['percent'] = (app_freq['tts_x']/app_freq['tts_y']) * 100

app_freq = app_freq.rename(columns={'tts_x': 'Sum App TTS', 'tts_y': 'Total Cat TTS'})

app_freq.head()
```

```python
app_freq.to_html('SumTTSAppsPerCluster.html')
```

#### Average TPS For Each App In Each Cluster

```python
app_freq = data.groupby(['app_name', 'model_predict', 'model_descrip']
                       )['tps'].sum().reset_index().sort_values(by=['model_descrip', 'tps'], 
                                                                      ascending=False) 

temp = app_freq.groupby('model_predict')['tps'].sum().reset_index()

app_freq = pd.merge(app_freq, temp, how='left', on='model_predict', copy=False)

app_freq['percent'] = (app_freq['tps_x']/app_freq['tps_y']) * 100

app_freq = app_freq.rename(columns={'tps_x': 'Sum App TPS', 'tps_y': 'Total Cat TPS'})

app_freq.head()
```

```python
app_freq.to_html('AvgTPSAppsPerCluster.html')
```

## Clustering in Three Dimensions

```python
sns.countplot(x = y_clusters, palette = 'YlGnBu')
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[y_clusters == 0,0],x[y_clusters == 0,1],x[y_clusters == 0,2], s = 40 , color = '#C9EEAF', label = "cluster 0")
ax.scatter(x[y_clusters == 1,0],x[y_clusters == 1,1],x[y_clusters == 1,2], s = 40 , color = '#377FA6', label = "cluster 1")
ax.scatter(x[y_clusters == 2,0],x[y_clusters == 2,1],x[y_clusters == 2,2], s = 40 , color = '#6EBBAD', label = "cluster 2")
ax.scatter(x[y_clusters == 3,0],x[y_clusters == 3,1],x[y_clusters == 3,2], s = 40 , color = '#393487', label = "cluster 3")

ax.set_xlabel('Sessions -->')
ax.set_ylabel('TTS -->')
ax.set_zlabel('DAU -->')
ax.legend()
plt.show()
```
